```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("/Users/arjun/Documents/Lab-Programs/5thSemLab/MachineLearning/Datasets/bank-data.csv")
```

```python
#Raw Measures Count, Values
print("Raw Measures Count, Values")
print(df["married"].values)
print(df["married"].count())
```

```python
#Binarization
print("Binarization")
threshold = 1
df["children"] = np.where(df["children"] > threshold, 1, 0)
print("Binarization")
print(df["children"].values)
```

```python
#Rounding
print("Rounding")
df["income_rounded"] = df["income"].round(decimals=2)
print(df["income_rounded"].values)
```

```python
#Interactions
print("Interactions")
df["interactions"] = df["income"] * df["save_act"]
print(df["interactions"].values)
```

```python
#Binning - Fixed Width Binnining
bin_width = 10
df["Income Bin"] = pd.cut(df['income'], bins=np.arange(0, df["income"].max() + bin_width,bin_width), right=False)
print(df["Income Bin"].values)
```

```python
#Binning - Quantile based binning
num_bins=3
df['Income Quantile Binning'] = pd.qcut(df['income'], q=num_bins, labels=False)
print("Binning-Quantile Based Binning", df['Income Quantile Binning'].values)
```

```python
# Mathematical Transformations - Log transform
print("# Mathematical Transformations - Log transform")
df["log_income"] = np.log(df["income"])
print(df["log_income"].values)
```

```python
# Mathematical Transformations - Box-Cox transform
print(" Mathematical Transformations - Box-Cox transform")
df["BOX-COX-income"], _ = stats.boxcox(df["income"])
print(df["BOX-COX-income"].values)
```
