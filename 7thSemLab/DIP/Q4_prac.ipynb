```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("./images/flower1.jpeg",0)
min_intensity = np.min(img)
max_intensity = np.max(img)
stretched_img = (((img - min_intensity)/(max_intensity - min_intensity)*255).astype(np.uint8))

hist_original = cv2.calcHist([img], [0], None, [256], [0,256])
hist_stretched = cv2.calcHist([stretched_img], [0],None, [256], [0, 256])
equalized_img = cv2.equalizeHist(img)
hist_equalized_img = cv2.calcHist([equalized_img], [0],None, [256], [ 0, 256])

plt.figure(figsize=(12,6))

plt.subplot(231)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(232)
plt.title("Streched Image")
plt.imshow(stretched_img)
plt.axis("off")

plt.subplot(233)
plt.title("Equalized Image")
plt.imshow(equalized_img)
plt.axis("off")


plt.subplot(234)
plt.title("Original Image Histogram")
plt.plot(hist_original)
plt.xlim([0, 256])

plt.subplot(235)
plt.title("Streched Image HIST")
plt.plot(stretched_img)
plt.xlim([0, 256])

plt.subplot(236)
plt.title("Equalized Image")
plt.plot(hist_equalized_img)
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
