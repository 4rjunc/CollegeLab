```python
from PIL import Image, ImageOps
```

```python
def display_negative(path):
    img = Image.open(path)
    print("Image")
    display(img)
    
    inverted_image = ImageOps.invert(img)
    print("Inverted Image")
    display(inverted_image)
    
    gray_scale = ImageOps.grayscale(img)
    print("Gray Scale Image")
    display(gray_scale)
    
    gray_scale_invert = ImageOps.invert(gray_scale)
    print("Gray Scale Invert Image")
    display(gray_scale_invert)

    binary_img = gray_scale.point(lambda p :  p > 128 and 255)
    display("Binary Image", binary_img)
```

```python
path = "./images/flower1.jpeg"
display_negative(path)
```
