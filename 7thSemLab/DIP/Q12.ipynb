```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./images/flower1.jpeg")

gblur = cv2.GaussianBlur(img, (3, 3), 0)
gray_scale = cv2.cvtColor(gblur, cv2.COLOR_BGR2GRAY)
edges = cv2.Laplacian(
    gray_scale, -1
)

plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(132)
plt.title("Gaussian Blur of Image")
plt.imshow(cv2.cvtColor(gblur, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(133)
plt.title("Laplacian Edge of Image")
plt.imshow(edges, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```
