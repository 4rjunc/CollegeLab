```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('./images/flower.jpg')


plt.subplot(241)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(242)
resized_image = cv2.resize(image, (200, 200))
plt.imshow(resized_image)
plt.axis("off")


plt.subplot(243)
rotate_image= cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotate_image)
plt.axis('off')

plt.subplot(244)
flipped_horizontal = cv2.flip(image, 1)
plt.imshow(flipped_horizontal)
plt.axis('off')


plt.subplot(245)
rotate_resized_image= cv2.rotate(resized_image, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotate_resized_image)
plt.axis('off')


plt.subplot(246)
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')


plt.subplot(247)
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)
plt.imshow(blurred_image)
plt.axis('off')

plt.tight_layout()
plt.show()
```
