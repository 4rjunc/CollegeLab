```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('./images/flower.jpg')

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

resized_image = cv2.resize(image, (200, 200))
plt.imshow(resized_image)
plt.axis("off")
plt.show()

rotate_image= cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotate_image)
plt.axis('off')
plt.show()

flipped_horizontal = cv2.flip(image, 1)
plt.imshow(flipped_horizontal)
plt.axis('off')
plt.show()

rotate_resized_image= cv2.rotate(resized_image, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotate_resized_image)
plt.axis('off')
plt.show()
```
