```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 

img = cv2.imread("./images/circle.jpeg")

gblur = cv2.GaussianBlur(img, (3,3), 0)

kernal = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

sharpening = cv2.filter2D(img, -1, kernel=kernal)

edges = cv2.Canny(image=gblur, threshold1=100, threshold2=100)

plt.figure(figsize=(9,9))

plt.subplot(221)
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')

plt.subplot(222)
plt.title("Edges Image")
plt.imshow(edges)
plt.axis('off')

plt.subplot(223)
plt.title("Sharpening Image")
plt.imshow(gblur)
plt.axis('off')

plt.show()
```
