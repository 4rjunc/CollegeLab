```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("./images/flower.jpg",0)
min_intensity = np.min(img)
max_intensity = np.max(img)

str_img = (((img- min_intensity)/(max_intensity - min_intensity)) * 255).astype(np.uint8)
hist_original = cv2.calcHist([img],[0],None,[256],[0, 256])
hist_stretched = cv2.calcHist([str_img],[0],None,[256],[0,256])

equalized_img = cv2.equalizeHist(img)
hist_equalized_img = cv2.calcHist([equalized_img],[0],None,[256],[0,256])

plt.figure(figsize=(12,8))

plt.subplot(231)
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')

plt.subplot(232)
plt.title("Stretched Image")
plt.imshow(str_img)
plt.axis('off')
 
```
